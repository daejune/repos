# Gini Impurity

$$ 1 - {P_{Surivived}} ^ 2 - {P_{Perish}} ^ 2 $$

## Good !
$$1 - ({100 \over 100}) ^2 -({0 \over 100})^2$$

## Bad!
$$1 - ({50 \over 100}) ^2 -({50 \over 100})^2$$

```python
import pandas as pd

train = pd.read_csv("titanic/train.csv")

print(train.shape)
train.head()
```

##  Visualize

### matplot
제일 기본적이고 많이 사용되지만 시각화를 위해 많은 코드가 필요함
### seaborn
matplot 기반으로 동작하는 시각화 모듈, 코드의 라인 수가 적은 것이 장점<br>
https://seaborn.pydata.org/

## Explore

```python
%matplotlib inline
import seaborn as sns
```

## Countplot
y축이 없음

```python
sns.countplot(data=train, x="Embarked")
```

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

## Barplot
y축이 있음: 수치  Data를 사용함

```python
sns.barplot(data=train, x="Pclass", y="Fare")
```

```python
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## Pointplot
barplot과 거의 동일한 문법으로 사용됨!!

X 축 Data가 서로 연관이 있을 때 주로 사용하고<br>
서로 연관이 없을 때는 barplot을 이용

```python
sns.pointplot(data=train, x="Pclass", y="Fare")
```

```python
sns.pointplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

## Lmplot
fit_reg: 추세선<br>
Excel에서의 분산형으로 뿌려주는 것과 동일함<br>
Data의 특정 영역을 Focusing 하기 좋음

```python
sns.lmplot(data=train, x="Age", y="Fare")
```

```python
sns.lmplot(data=train, x="Age", y="Fare", fit_reg=False)
```

```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

```python
low_fare = train[train["Fare"] < 500]
sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

```python
low_fare = train[train["Fare"] < 100]
sns.lmplot(data=low_fare, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

## Distplot
data의 Column을 직접 지정해줘야함<br>
Excel에서 구간 영역을 나눈 후 히스토그램 분석<br>
Bar는 Whether to plot a (normed) histogram<br>
Line은 Whether to plot a gaussian kernel density estimate

```python
sns.distplot(train["Fare"])
```

```python
sns.distplot(low_fare["Fare"])
```

```python
sns.distplot(low_fare["Fare"], hist=False)
```
