```python
import pandas as pd
# Load Training Data Set
train = pd.read_csv("train.csv", index_col = "PassengerId")
# Load Test Data Set
test = pd.read_csv("test.csv", index_col = "PassengerId")
# Load Summit Data Set
submit = pd.read_csv("gender_submission.csv", index_col = "PassengerId")
```

```python
test.head()
```

## Preprocessing

### Sex Encoding (True / False or 1 / 0 Case)

```python
train.loc[train["Sex"] == "male", "SexEncode"] = 0
train.loc[train["Sex"] == "female", "SexEncode"] = 1

print (train.shape)
train.head()
```

```python
test.loc[test["Sex"] == "male", "SexEncode"] = 0
test.loc[test["Sex"] == "female", "SexEncode"] = 1

print (test.shape)
test.head()
```

### Fall in missing Fare

```python
train[train["Fare"].isnull()]
```

```python
train.loc[train["Fare"].isnull(), "Fare"] = 0

print(train.shape)
train.head()
```

```python
train[train["Fare"].isnull()]
```

```python
test[test["Fare"].isnull()]
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = 0

print(test.shape)
test.head()
```

```python
test[test["Fare"].isnull()]
```

### Embarked Encoding (One Hot Encoding Case)

```python
# C == 0, S == 1, Q == 2
# 2 * S == Q?
# S + S == Q?

# One Hot Encoding
# True == 1, False == 0
# C == [True, False, False]
# S == [False, True, False]
# Q == [False, False, True]
```

```python
train["EmbarkedC"] = train["Embarked"] == "C"
train["EmbarkedS"] = train["Embarked"] == "S"
train["EmbarkedQ"] = train["Embarked"] == "Q"

print (train.shape)
train[["Embarked", "EmbarkedC", "EmbarkedS", "EmbarkedQ"]].head()
```

```python
test["EmbarkedC"] = test["Embarked"] == "C"
test["EmbarkedS"] = test["Embarked"] == "S"
test["EmbarkedQ"] = test["Embarked"] == "Q"

print (test.shape)
test[["Embarked", "EmbarkedC", "EmbarkedS", "EmbarkedQ"]].head()
```

```python
feature_names = ["Pclass", "SexEncode", "Fare", "EmbarkedC", "EmbarkedS", "EmbarkedQ"]
feature_names
```

```python
train_x = train[feature_names]

print(train_x.shape)
train_x.head()
```

```python
test_x = test[feature_names]

print(test_x.shape)
test_x.head()
```
