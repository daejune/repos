## Column Add & Delete

```python
new_train = train
new_train["Category"]
```

```python
new_train["Category"] = "Titanic"
new_train.head()
```

```python
new_train["Id"] = range(890)
```

```python
new_train["Id"] = range(892)
```

```python
# 갯수를 정확인 맞춰야한다.
new_train["Id"] = range(891)
new_train
```

```python
new_train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
new_train[["SibSp", "Parch", "FamilySize"]].head()
```

```python
new_train["France"] = new_train["Embarked"] == "C"
new_train["UK"] = new_train["Embarked"] != "C"
new_train[["Embarked", "France", "UK"]].head()
```

```python
new_train.loc[new_train["Embarked"] == "C", "Nationality"] = "France"
new_train[["Embarked", "Nationality", "France", "UK"]].head()
```

```python
new_train.loc[new_train["Embarked"] != "C", "Nationality"] = "UK"
new_train[["Embarked", "Nationality", "France", "UK"]].head()
```
