```python
# load data
import pandas as pd
train = pd.read_csv("bike_sharing/train.csv")
print (train.shape)
test = pd.read_csv("bike_sharing/test.csv")
print (test.shape)
```

```python
%matplotlib inline
import seaborn as sns
```

```python
feature_names = ["datetime", "season", "holiday", "workingday", "weather", "temp", "atemp", "humidity", "windspeed"]

is_train = train[feature_names].copy()
is_train["train"] = 1
is_test = test[feature_names].copy()
is_test["train"] = 0

is_train.to_csv("bike_sharing/train_s.csv", index=False)
is_test.to_csv("bike_sharing/test_s.csv", index=False)
```
